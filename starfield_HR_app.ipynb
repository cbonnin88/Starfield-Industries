```python
class Employee:
  def __init__(self, emp_id, name, position, salary):
    self.emp_id = emp_id
    self.name = name
    self.position = position
    self.salary = salary
    self.performance_score = None
    self.Leaves_taken = 0

  def update_leave(self, days):
    self.Leaves_taken += days
    return f"{self.name} has taken {days} of leave. Total leave taken: {self.Leaves_taken}"

  def update_performance(self,score):
    self.performance_score = score
    return f"{self.name}'s performance score updated to {score}."

  def get_details(self):
    return f"ID: {self.emp_id},\nName: {self.name}\nPosition: {self.position}\nSalary: {self.salary}\nLeave Taken: {self.Leaves_taken}\nPerformance: {self.performance_score}"

class HRSystem:
  def __init__(self):
    self.employees = {}

  def add_employee(self):
    emp_id = input("Enter Employee ID: ")
    name = input("Enter Employee Name: ")
    position = input("Enter Employee Position: ")
    salary = float(input("Enter Salary: "))
    self.employees[emp_id] = Employee(emp_id, name, position, salary)
    print(f"Employee {name} added successfully.")

  def remove_employee(self):
    emp_id = input("Enter Employee ID to remove: ")
    if emp_id in self.employees:
      del self.employees[emp_id]
      print("Employee successfully removed.")
    else:
      print("Employee not found")

  def record_Leave(self):
    emp_id = input("Enter Employee ID: ")
    if emp_id in self.employees:
      days = int(input("Enter number of leave days: "))
      print(self.employees[emp_id].update_leave(days))
    else:
      print("Employee not found")

  def record_performance(self):
    emp_id = input("Enter Employee ID: ")
    if emp_id in self.employees:
      score = float(input("Enter performance score(1-10): "))
      print(self.employees[emp_id].update_performance(score))
    else:
      print("Employee not found")

  def display_employees(self):
    if not self.employees:
      print("No employees in the system")
      return
    for emp in self.employees.values():
      print(emp.get_details())

  def run(self):
     while True:
      print("\nHR Management System")
      print("1. Add an Employee")
      print("2. Remove Employee")
      print("3. Record Leave")
      print("4. Record Performance")
      print("5. Display All Employees")
      print("6. Exit")

      choice = int(input("Enter your choice: "))

      if choice == 1:
        self.add_employee()
      elif choice == 2:
        self.remove_employee()
      elif choice == 3:
        self.record_Leave()
      elif choice == 4:
        self.record_performance()
      elif choice == 5:
        self.display_employees()
      elif choice == 6:
        print("Exiting the HR System")
        break

      else:
          print('Invalid choice, please try again')

if __name__ == '__main__':
    starfield_system = HRSystem()
    starfield_system.run()
```
