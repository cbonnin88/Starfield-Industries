```python
matplotlib.rc("figure",figsize=(20,10))
```

```python
df_starfield['Salary Band'].value_counts().plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Salary Band counts')
plt.show()
```

```python
df_starfield["Performance"].value_counts()
```

```python
df_starfield['Performance'].value_counts().plot(kind='bar')
plt.xticks(rotation=45)
plt.title("Performance Level Count")
plt.show()
```

```python
gender_labels = df_starfield['Gender'].unique()
gender_counts = df_starfield['Gender'].value_counts()

plt.pie(gender_counts,
      labels=gender_labels,
      autopct='%1.0f%%',
      startangle=90)

plt.title('Gender Distribution')
plt.show()
```
